```python
import os
#os.mkdir("Data/spoken_numbers")
os.mkdir("Data/spoken_numbers/recordings")
```

```python
class _TTS:

    engine = None
    rate = None
    def __init__(self):
        self.engine = pyttsx3.init()


    def start(self,text_,filename):
        rate = self.engine.getProperty('rate')   # getting details of current speaking rate
        self.engine.save_to_file(text, f'Data/spoken_numbers/recordings/{filename}')
        self.engine.runAndWait()
        return rate
```

```python
for number in tqdm(range(1000)):
    text = num2words(number)
    text = text.replace('-', ' ')
    filename = text.replace(' ', '_') + '.wav'
    
    
    engine = _TTS()
    rate = engine.start(text,filename)
    del(engine)
```

```python
data = []
for number in range(1000):
    text = num2words(number)
    text = text.replace('-', ' ')
    filename = text.replace(' ', '_') + '.wav'
    data.append({'text': text, 'filename': filename})
```

```python
data = pd.DataFrame(data)
data.to_csv('Data/spoken_numbers/data.csv', index=False)


```
