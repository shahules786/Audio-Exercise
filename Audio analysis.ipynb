## Basic analysis

```python
ipd.Audio("Data/chunked/a.wav")
```

```python
sound = sound/ (2**15)
```

## FFT analysis

```python
X_fft = np.fft.rfft(sound)
threshold = np.abs(X_fft)/10000
X_fft[X_fft<threshold] =0 

freq_bins = np.fft.fftfreq(len(X_fft),d=1/sampleFreq)

plt.plot(freq_bins,np.abs(X_fft))
plt.xlim(-1000,22100)
```

```python
plt.plot(freq_bins[:100],np.abs(X_fft)[:100])
```

### Phase spectra

```python
plt.plot(freq_bins,np.angle(X_fft) * 180/np.pi)
plt.xlim(-100,100)
```
